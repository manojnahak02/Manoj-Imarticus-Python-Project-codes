# Discretization and Binning

```python
passangerage=Titanic_df['Age'] #created duplicate column for age varable
passangerage = passangerage.dropna() #deleting the missing values
Bins = [0,15,21,60,passangerage.max()] #creating buckets for age
Binlabels=['Children','Adolescents','Adult','Senior'] #specified names for the buckets
categories=pd.cut(passangerage,Bins,labels=Binlabels)
print(categories.value_counts())
```

```python
#concatination two dataframes or series

new_df=pd.concat([passangerage,categories], axis = 1) #axis =1 for column wise concatination
new_df.head(12)
```

# Droping variables

```python
#Generic data sets where loss is in money, Drop a varible if Thres MV > 50%
#Sensitive data sets where there is loss on Human, Drop a varible if Thres MV > 75%

half_count = len(Titanic_df)/2

Titanic_df = Titanic_df.dropna(thresh = half_count, axis = 1) #drop a variable if MV is > 50%
print(Titanic_df.isnull().sum())

"""
half_count = 0.75 * len(Titanic_df)
print(half_count)
""""
```

```python
Titanic_df = Titanic_df.drop(['Ticket', 'Fare'], axis = 1)
Titanic_df.head()
```

# Treating the missing values

```python
#handling MV in embarked column

#inplace= true is used for assigning values to the missing values, it is faster and effecient

Titanic_df['Embarked'].fillna(Titanic_df['Embarked'].mode()[0], inplace=True)

#or another way to replace the MV

#Titanic_df['Embarked']=Titanic_df['Embarked'].fillna(Titanic_df['Embarked'].mode()[0])
print(Titanic_df.isnull().sum())
```

```python
Titanic_df['Embarked'].mode()
```

```python
#handling MV in Age column
Titanic_df['Age'].fillna(round(Titanic_df['Age'].mean(),0), inplace=True)
print(Titanic_df.isnull().sum())
#print(Titanic_df.head(10))
```
